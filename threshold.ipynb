```python
#import libraries
import csv
import pandas as pd
import shutil
import os
```

```python
#check which folder does this ipynb file is
print(os.getcwd())
```

```python
#Read the ao3 stories general metadata
data = pd.read_csv('mobius_ao3_works_z1.csv', encoding = 'latin1')
df = pd.DataFrame(data, columns=['id', 'fandom', 'title', 'authors', 'date_updated', 'hits', 'language', 'words'])
df
```

```python
#here is the authors who wrote more than 10 books
df_new = df.loc[(df['words'] >= 2000) & (df['hits'] >= 500)  & (df['language'] == 'English')]
selection = df_new[df_new.groupby('authors')['authors'].transform('size').gt(10)]
selection

```

```python
#get the story-id list
story_id = selection['id'].tolist()
```

```python
#save selected story-id into the csv file
dict = {'id': story_id}  
       
id_df = pd.DataFrame(dict) 
    
# saving the dataframe 
id_df.to_csv('ao3_select_storyid.csv') 
```

```python
#这里考虑到数字钱leading zero的问题，通过xlsx进行处理，处理后文件为text.xlsx
#in order to keep the leading zero, I use xlsx to fix the problem. The file is text.xlsx"
ids = pd.read_excel('test.xlsx', dtype='object')
ids_df = pd.DataFrame(ids, columns=['id'])
ids_df
```

```python
story_id_w0 = ids_df['id'].tolist()
```

```python
#get the dataframe of all the stories that fits the creteria
id_selection_df = selection[selection['id'].isin(story_id)]
id_selection_df
```

```python
#get the author name
author_list = id_selection_df["authors"].unique()
for i in author_list:
    print(i)
print(len(author_list))
```

```python
#this step is to change the story-id with leading zeros into strings
story_id_w0_str = [str(x) for x in story_id_w0]
```

```python
#This step is to move the select story from the corpus to another folder
source_dir = '/Users/zeyu/Desktop/data/mobius/'
filename_lst = []
for filename in os.listdir(source_dir):
    filename_lst.append(os.path.basename(filename)) 
#first to check the number of files in the original folder
print(len(filename_lst))
```

```python
#this step is to select the stories that are selected and need to be moved.
story_id_w0_str_ext = [i + '.txt' for i in story_id_w0_str]
filename_lst_select = set(story_id_w0_str_ext).intersection(filename_lst)
print(len(filename_lst_select))
```

```python
#this step is to move the selected stories into a new folder from the original corpus folder
source_dir = '/Users/zeyu/Desktop/data/mobius/'
target_dir = '/Users/zeyu/Desktop/data/mobius_select'
for filename in filename_lst_select:
    source_filename = os.path.join(source_dir, filename)
    target_filename = os.path.join(target_dir,filename )
    shutil.move(source_filename, target_filename)
```

```python
#read the MTLD score dataframe
mtld_df= pd.read_csv('mobius_select_mtld.csv')#, dtype='object')
mtld_df = pd.DataFrame(mtld_df, columns=['mtld','id'])
mtld_df
```

```python
#merge the MTLD score dataframe with the selected stories metadata pd frame
id_mtld_meta_df = pd.merge(id_selection_df, mtld_df, on='id')
id_mtld_meta_df
```

```python
#check the "author" details for the remaining stories
id_mtld_meta_df['authors'].value_counts()
```

```python
#make sure each author have at least 10 stories
N=10
id_mtld_meta =id_mtld_meta_df[id_mtld_meta_df.groupby('authors')['authors'].transform('count')>N].copy()
id_mtld_meta
```

```python
#double check, make sure that each author have at least 10 stories in this data frame
id_mtld_meta['authors'].value_counts()
```

```python
#save the dataframe into csv files
id_mtld_meta_df.to_csv('id_mtld_meta_df.csv')
```
